# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
pd.DataFrame(collection_companies.find({'name':'Babelgum'},{'_id':0,'name':1}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
#En sort() -1 es descendente, 1 es ascendente
pd.DataFrame(collection_companies.find({'number_of_employees':{"$gt":5000}}).limit(20).sort("number_of_employees",1)).head(3)
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
#Pones el $and y en seguida una lista de dicts con todas las condiciones a cumplir
pd.DataFrame(collection_companies.find({'$and':[{"founded_year":{"$gte":2000}},{"founded_year":{"$lte":2005}}]},{"_id":0,"name":1,"founded_year":1})).head(3)
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
#Cuando es un campo dentro de otro se separan con un punto ej: "ipo.valuation_amount"
pd.DataFrame(collection_companies.find({'$and':[{"ipo.valuation_amount":{"$gt":100000000}},{"founded_year":{"$lt":2010}}]},{"_id":0,"name":1,"ipo":1})).head(3)
```

```python
pd.json_normalize(collection_companies.find({'$and':[{"ipo.valuation_amount":{"$gt":100000000}},{"founded_year":{"$lt":2010}}]},{"_id":0,"name":1,"ipo":1}).sort("ipo.valuation_amount",-1)).head(3)
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
pd.DataFrame(collection_companies.find({'$and':[{"number_of_employees":{"$lt":1000}},{"founded_year":{"$lt":2005}}]}).sort('number_of_employees').limit(10)).head(3)
```

### 6. All the companies that don't include the `partners` field.

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
pd.DataFrame(collection_companies.find({'$and':[{"founded_month":{"$gt":5}},{"founded_month":{"$lte":12}}]}).limit(1000)).head(3)
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
#$expr compara con el operador logico q le des
#En este caso pide q deadpooled... se gt founded +3
#$add hace una suma entre el valor d un campo y el numero q pases $add:[<campo>,<# a sumar>]
pd.DataFrame(collection_companies.find({'$and':[{'$expr':{'$gt':['$deadpooled_year',{'$add':['$founded_year',3]}]}},{'founded_year':{'$gte':0}}]}))
```

### 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
pd.DataFrame(collection_companies.find({"$and":[{"founded_year":{"$lt":2000}},{"acquisition.price_amount":{"$gt":10000000}}]})).head()
```

### 14. All the companies that have been acquired after 2015, order by the acquisition amount, and retrieve only their `name` and `acquisiton` field.
